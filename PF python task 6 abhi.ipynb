Develop a graph which represents the public transport network of a city of your choice. Some cities have extensive public transport networks. In such case, represent minimum 3 lines with minimum 4 stations on each line. When visualising the network, use different colours for the different lines and their corresponding stations. Provide attributes to the edges which correspond to the distances between stations and visualise them. If the actual distances between the stations are not available, approximate them by using online map services. Visualise the names of the stations.

For this task, I'm using 3 major train lines in london,UK. Jubilee Line, District line, Central line.

```python
# Installing libraries
import networkx as nx # library for creating network
import matplotlib.pyplot as plt # library for ploting the graph
import pandas as pd # Library for extension of data.

# Create a graph object
MyGraph = nx.Graph()

# Adding nodes

# Grey Line
MyGraph.add_node('1', npos=(20, 10), ccn='#808080')
MyGraph.add_node('2', npos=(55, 10), ccn='#808080')
MyGraph.add_node('3', npos=(100, 10), ccn='#808080')
MyGraph.add_node('4', npos=(128, 25), ccn='#808080')
MyGraph.add_node('5', npos=(128, 57), ccn='#808080')
MyGraph.add_node('6', npos=(100, 68), ccn='#808080')

# yellow Line
MyGraph.add_node('7', npos=(100, 62), ccn='#FFFF00')
MyGraph.add_node('8', npos=(80, 48), ccn='#FFFF00')
MyGraph.add_node('9', npos=(65, 39), ccn='#FFFF00')
MyGraph.add_node('10', npos=(50, 32), ccn='#FFFF00')
MyGraph.add_node('11', npos=(35, 25), ccn='#FFFF00')
MyGraph.add_node('12', npos=(20, 18), ccn='#FFFF00')

# Green Line
MyGraph.add_node('13', npos=(130, 100), ccn='#228B22')
MyGraph.add_node('14', npos=(130, 90), ccn='#228B22')
MyGraph.add_node('15', npos=(130, 80), ccn='#228B22')
MyGraph.add_node('16', npos=(100, 65), ccn='#228B22')
MyGraph.add_node('17', npos=(55, 44), ccn='#228B22')
MyGraph.add_node('18', npos=(15, 35), ccn='#228B22')

# Connect nodes
# Grey Line
MyGraph.add_edge('1', '2', cce='#808080',weight = 3.1)
MyGraph.add_edge('2', '3', cce='#808080',weight = 3.4)
MyGraph.add_edge('3', '4', cce='#808080',weight = 2.8)
MyGraph.add_edge('4', '5', cce='#808080',weight = 1.7)
MyGraph.add_edge('5', '6', cce='#808080',weight = 1.8)

# yellow Line
#MyGraph.add_edge('7', '7', cce='#FF4500')
MyGraph.add_edge('7', '8', cce='#FFFF00',weight = 1.3)
MyGraph.add_edge('8', '9', cce='#FFFF00',weight = 0.8)
MyGraph.add_edge('9', '10', cce='#FFFF00',weight = 0.5)
MyGraph.add_edge('10', '11', cce='#FFFF00',weight = 0.6)
MyGraph.add_edge('11', '12', cce='#FFFF00',weight = 0.3)

# Green Line
#MyGraph.add_edge('L', 'M', cce='#0000FF')
MyGraph.add_edge('12', '13', cce='#228B22',weight = 1.3)
MyGraph.add_edge('13', '14', cce='#228B22',weight = 1.7)
MyGraph.add_edge('14', '15', cce='#228B22',weight = 1.7)
MyGraph.add_edge('15', '16', cce='#228B22',weight = 2)
MyGraph.add_edge('16', '17', cce='#228B22',weight = 0.6)

# Extract attributes from the graph to dictionaries
pos = nx.get_node_attributes(MyGraph, 'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')
edgewt = nx.get_edge_attributes(MyGraph, 'weight')

# Place the dictionary values in lists
NodeList = list(nodecolour.values())
EdgeList = list(edgecolour.values())
EdgeWeight = list(edgewt.values())
# Set the size of the figure
plt.figure(figsize=(10, 10))

# Display the names of the stations
# Jubilee line
plt.text(23, 8, s='Westminister', rotation=0)
plt.text(58, 8, s='Waterloo', rotation=0)
plt.text(103, 9, s='Southwark', rotation=0)
plt.text(109, 24, s='London Bridge', rotation=0)
plt.text(113, 56, s='Canada Water', rotation=0)
plt.text(87, 65, s='Canary Warf', rotation=0)

plt.text(93, 53, s='South kensington', rotation=0) 
plt.text(58, 38, s='Victoria', rotation=0)
plt.text(58, 30, s='Westminister', rotation=0)
plt.text(58, 22, s='Embankment', rotation=0)
plt.text(58, 14, s='Cannon Street', rotation=0)
plt.text(58, 14, s='Monument', rotation=0)

plt.text(113, 99, s='South kensington', rotation=0)
plt.text(113, 89, s='Victoria', rotation=0)
plt.text(119, 80, s='Westminister', rotation=0)
plt.text(42, 44, s='Embankment', rotation=0)
plt.text(13, 39, s='Cannon Street', rotation=0)
plt.text(13, 39, s='Monument', rotation=0)

# Draw the nodes and the edges
nx.draw_networkx(MyGraph, pos, node_color=NodeList)
nx.draw_networkx_edges(MyGraph, pos, edge_color=EdgeList)
nx.draw_networkx_edge_labels(MyGraph, pos, edge_labels=edgewt)
# Visualise the graph
plt.show()
```
